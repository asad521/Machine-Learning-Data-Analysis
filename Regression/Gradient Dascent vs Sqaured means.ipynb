```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

def grd(x,y):
    m_cur=0
    b_cur=0
    iteration=10000
    ss=0.001 # step size, iteration step, learning rate3
    n=len(x)
    for i in range(iteration):
                
        y_predicted=x*m_cur+b_cur
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])
        dm= -(2/n)*sum(x*(y-y_predicted))
        db= -(2/n)*sum(y-y_predicted)
        m_cur=m_cur- ss*dm
        b_cur=b_cur-ss*db
        print(f"Value of slope  is:{m_cur}, Value of intercept is:{b_cur}\n Value"
              +f"is cost is {cost}")
        plt.scatter(x,y, color='blue')  #plot previous plot
        plt.plot(x,y_predicted, color='red')
#         plt.ylim([0,20])
#         plt.xlim([0,50])
# x = np.array([12,23,33,44,54])
# y = np.array([5,7,9,11,13])
x = np.array([2,3,5,13,8,16,11,1,9])
y = np.array([15,28,42,64,50,90,58,8,54])
grd(x,y)
##There is one confusion that both m and b start from 0 values  and then gradually increase with same stop size. 
#Then question arrise 1)How b values get change after 1000 iteration. Both values of m and b should be same.
# 2)B and m values should also not be same, because it does not necessary that m and y will be zero at slope
###########
# There values get change because both calculation the dereviate of line equation seprative and then change is subtracted 
# from learning rate and step size. Means new value  of slop and b is depends on previous values- stepsize and derivative
#One more thing, above algorithem will not tell which value of m and b is perfect .you will have to calculate cost function
# on each iteration(each value of m and b) to check which value of m and b give minimum cost function.
########+++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++++##########
# WAHT WE ARE DOING ACUTALLY.
# ACUTALY WE ARE Derrivatin the cost function with respect to m and b one by one. 
# means we are finding "change" of cost function with repsect to m and b
# then we use those changes to find value of m and b 
# then we put those value in cost function of equation to check whether cost function reduces or not
# and keep repeating these steps
```
