```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
tips=sns.load_dataset("tips")
iris=sns.load_dataset("iris")
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips)
```

```python
sns.jointplot(x='sepal_length',y='sepal_width', data=iris)
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='reg') #adding reggsion line  on scatterand densition function on history
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='hex') #changing  data to hexagon shape
```

```python
sns.jointplot(x='total_bill',y='tip', data=tips, kind='kde') #kde of total_bill and tips
```

```python
from scipy.stats import spearmanr
sns.jointplot(x='size',y='total_bill', data=tips, stat_func=spearmanr) #given spearman 
```
