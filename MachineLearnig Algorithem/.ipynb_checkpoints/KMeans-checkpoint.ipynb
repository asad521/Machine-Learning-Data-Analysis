```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as  plt
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.cluster import KMeans
```

```python
df=pd.read_excel("income.xlsx", Sheet="Book1")
df
```

```python
plt.scatter(df.Age,df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df["cluster"]=y_predicted
df
```

```python
km.cluster_centers_
```

```python
c1=df[df["cluster"]==1]
c2=df[df["cluster"]==2]
c0=df[df["cluster"]==0]
plt.scatter(c0["Age"],c0["Income($)"],color='red', label='Cluster 1')
plt.scatter(c1["Age"],c1["Income($)"],color='blue', label='Cluster 2')
plt.scatter(c2["Age"],c2["Income($)"],color='yellow', label='Cluster 3')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Clusters")
```

```python
# We can see there yellow cluster is not possible. Because X have very lmited ranger as compared to income.
# Income is thousand.So we need to scale featuer properly
```

```python
scaler=MinMaxScaler()
scaler.fit(df[["Income($)"]])
df["Scaled_Income($)"]=scaler.transform(df[["Income($)"]])
scaler.fit(df[["Age"]])
df["Scaled_Age"]=scaler.transform(df[["Age"]])
df
```

```python
km=KMeans(n_clusters=3)
```

```python
# y_predicted=km.fit_predict(df[['Scaled_Age']],df[['Scaled_Income($)']]) #tHIS IS fALSE.wILL GIVE WRONG RESULT as per my exp
y_predicted=km.fit_predict(df[['Scaled_Age','Scaled_Income($)']]) #tHIS IS CORRECT 
y_predicted
```

```python
df["Scaled_After_Cluster"]=y_predicted
df
```

```python
c1=df[df["Scaled_After_Cluster"]==1]
c2=df[df["Scaled_After_Cluster"]==2]
c0=df[df["Scaled_After_Cluster"]==0]
plt.scatter(c0["Scaled_Age"],c0["Scaled_Income($)"],color='red', label='Cluster 1')
plt.scatter(c1["Scaled_Age"],c1["Scaled_Income($)"],color='blue', label='Cluster 2')
plt.scatter(c2["Scaled_Age"],c2["Scaled_Income($)"],color='yellow', label='Cluster 3')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Clusters")
```

```python
km.cluster_centers_
```

```python
#plot centroied as well
c1=df[df["Scaled_After_Cluster"]==1]
c2=df[df["Scaled_After_Cluster"]==2]
c0=df[df["Scaled_After_Cluster"]==0]
plt.scatter(c0["Scaled_Age"],c0["Scaled_Income($)"],color='red', label='Cluster 1')
plt.scatter(c1["Scaled_Age"],c1["Scaled_Income($)"],color='blue', label='Cluster 2')
plt.scatter(c2["Scaled_Age"],c2["Scaled_Income($)"],color='yellow', label='Cluster 3')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Income')
plt.title("Clusters")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='+',label="Centroid")
```

```python
#Now we will see Elbow plot method.
k_range=range(1,11)
sse=[] #sum of sqaure error
for k in k_range:
    km=KMeans(n_clusters=k)
    km.fit(df[['Scaled_Age','Scaled_Income($)']]) #tHIS IS CORRECT 
    sse.append([km.inertia_])#km.inertia_ give sum of square error
```

```python
sse
```

```python
plt.xlabel("Number of K")
plt.ylabel("Sum of Square Error")
plt.plot(k_range,sse)
```
