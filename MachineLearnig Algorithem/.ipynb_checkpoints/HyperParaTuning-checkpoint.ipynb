```python
# #Problem with above approach is that we have to write each line with for different parameter.
# #Solution is to write for loopm
# list=["linear",'sigmoid','rbf']
# cv=[10,20,30]
# for par in list:
#     for c in cv:
#         print(f"kval:{par}  Cvalue:{c}")
#         print(cross_val_score(SVC(kernel=par,C=c,gamma='auto'),Xtrain,ytrain,cv=5))
#         print("")
#=====same as below=====
#Problem with above approach is that we have to write each line with for different parameter.
#Solution is to write for loopm
list=["linear",'sigmoid','rbf']
cv=[10,20,30]
avg_score={}
for par in list:
    for c in cv:
        score=(cross_val_score(SVC(kernel=par,C=c,gamma='auto'),Xtrain,ytrain,cv=5))
        avg_score[par + "_" +str(c)]=np.average(score) #assign to dictionary, keys=data
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#dictionary is different paramter=keys and therir required values =data
# class sklearn.model_selection.GridSearchCV(estimator, param_grid, *, scoring=None, n_jobs=None, 
# iid='deprecated', refit=True, cv=None, verbose=0, pre_dispatch='2*n_jobs', error_score=nan, return_train_score=False)
clf=GridSearchCV(SVC(gamma='auto'),{
    'C':[10,20,30],
    'kernel':["rbf","linear","sigmoid"]  
},cv=10,return_train_score=False)
```

```python
clf.fit(iris.data,iris.target)
```

```python
#cls is ready , now we can check differnt parameters of our results
clf.cv_results_  #result of classifier based on our conditions
```

```python
results=pd.DataFrame(clf.cv_results_ ) #Converting the results in dataframe
results
```

```python
results[["param_C","param_kernel","mean_test_score"]] #Only three cloumns are important,therfore seeing it
```

```python
#Above  code is GridSearchCV for one model .
#To select best model, we will do the follwing.define dictionary of models
model_params={
    'svm': {
     'model':SVC(gamma='auto'),
    'params':{
    'C':[1,10,20],
    'kernel':["rbf","linear","sigmoid"]
    }
    },
   'random_forest':{
    'model':RandomForestClassifier(),
    'params':{
 'n_estimators':[1,5,10],
    }
    },
    'logisticregression':{
    'model':LogisticRegression(solver='liblinear',multi_class='auto'),
    'params':{
     'C':[1,5,10],
    }
    },
    }
```

```python
scores=[]
for model_name,mp in model_params.items():
    clf=GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(iris.data,iris.target)
    scores.append({
       'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
         })
```

```python
df=pd.DataFrame(scores,columns=["model","best_score","best_params"])
```
