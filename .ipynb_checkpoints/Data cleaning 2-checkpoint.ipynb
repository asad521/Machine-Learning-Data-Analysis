```python
# for all columns  we will do fllowing
def cat_all_col_data(var):
    return pd.concat([df[var].value_counts()/df.shape[0]*100,
         finaldf[var].value_counts()/finaldf.shape[0]*100], axis=1, keys=[var+'_org',var+'_final'])
    
    
    
```
