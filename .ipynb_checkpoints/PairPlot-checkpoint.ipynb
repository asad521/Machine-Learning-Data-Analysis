```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

```python
tips=sns.load_dataset("tips")
iris=sns.load_dataset("iris")
```

```python
sns.pairplot(iris) #showing relationships between columns. Diagnola shows corelation of column to iteself
```

```python
sns.pairplot(iris ,hue="species") #passing hue
```

```python
sns.pairplot(iris ,hue="species", markers=['o','+','D']) #MARKER DIFFERENT
```

```python
sns.pairplot(iris ,hue="species", vars=['petal_length','sepal_width']) 

#It is same as pairgrid .you can also chose x and y axis by vars
```

```python
sns.pairplot(iris ,hue="species",kind='reg') # adding regression line 
```

```python
sns.pairplot(iris ,hue="species", diag_kind='kde') # chaning daiganl to kde kind
```
