```python
 days=[1,2,3,4,5,6,7]
```

```python
max_t=[34,38,30,39,31,33,29]
```

```python
min_t=[23,24,28,18,23,11,10]
```

```python
avg_t=[30,28,22,24,30,16,50]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather in Pakistan")
plt.plot(days,max_t, label='max')
plt.plot(days,min_t, label='min')
plt.plot(days,avg_t, label= 'vg')
plt.legend(loc= "best", shadow="true", fontsize="large")
plt.grid()
plt.axes()
```

```python
import numpy as np
```

```python
company=['Abc','Def','Meezan','Hilal']
```

```python
revenue=[234,334,66,63]
profit=[121,182,190,101]
```

```python
X = np.arange(len(company))
plt.bar(X + 0.00 , revenue, color = 'b', width = 0.25)
plt.bar(X + 0.25, profit, color = 'g', width = 0.25)
plt.xticks(X,company)
plt.xlabel('Company', size=14)
plt.ylabel('Revenue' ,size=14)
plt.title('US Tech Stock', size=15)
```
