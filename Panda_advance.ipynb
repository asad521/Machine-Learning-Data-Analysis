```python
###Merge ## merge two dataframe based on column
df1=pd.DataFrame({
    "city":["Mumbai","Karachi","Lahore"],
    "Humidity":[34,45,99],
    "temperature":[23,18,37],
})##0 for mum,1, karachi, 2 lahore
df1
df2=pd.DataFrame({
    "city":["Mumbai","Lahore","Karachi"],
    "Windspeed":[4,6,9],
    "Suntime":[33,14,8],
})# alinging ciities
df2
```

```python
# mERGE WILL MERGE TWO datafarame with same number of values. Means it will intersection of 2df based on city
df3=pd.merge(df1,df2,on="city") #No need to set index in original frame #Here you do not pass list as oppsed to concat
df3
```

```python
###Merge ## merge two dataframe based on column
df1=pd.DataFrame({
    "city":["Mumbai","Karachi"],
    "Humidity":[34,45],
    "temperature":[23,18],
})##0 for mum,1, karachi, 2 lahore
df1
df2=pd.DataFrame({
    "city":["Mumbai","Lahore","Karachi","SanFransisco"],
    "Windspeed":[4,6,9,8],
    "Suntime":[33,14,8,9],
})# alinging ciities
df2
# mERGE WILL MERGE TWO datafarame with same number of values. Means it will intersection of 2df based on city
df3=pd.merge(df1,df2,on="city", how="outer", indicator=True) # out will union for two dataframe.default is inner which is intersection
df3
#indicator just showing  where indivature values comes form
```
