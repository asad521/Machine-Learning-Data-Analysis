```python
#Pair plots

x=sns.PairGrid(data=iris) #It is comapring the columns with columns having numerical values
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(data=tips) #It is comapring the columns with columns having numerical values
x=x.map(plt.scatter)
```

```python
x=sns.PairGrid(data=iris) #It is comapring the columns with columns having numerical values
x=x.map_diag(plt.hist)# This function will only show diagonal in historgram form
```

```python
x=sns.PairGrid(data=iris) #It is comapring the columns with columns having numerical values
x=x.map_diag(plt.hist)# This function will only show diagonal in historgram form
x=x.map_offdiag(plt.scatter)# This is neccessary to show all other graph as well
```

```python
x=sns.PairGrid(data=iris, hue="species") #Including the another variable by hue. Will show in color form
x=x.map_diag(plt.hist)# This function will only show diagonal in historgram form
x=x.map_offdiag(plt.scatter)# This is neccessary to show all other graph as well
```

```python
x=sns.PairGrid(data=iris, hue="species") #Including the another variable by hue. Will show in color form
x=x.map_diag(plt.hist)# This function will only show diagonal in historgram form
x=x.map_offdiag(plt.scatter)# This is neccessary to show all other graph as well
x=x.add_legend()# To show legend
```

```python
x=sns.PairGrid(data=iris, hue="species") #Including the another variable by hue. Will show in color form
x=x.map_diag(plt.hist, histtype='step', linewidth=4)# changing histtype and line width of hist
x=x.map_offdiag(plt.scatter)# This is neccessary to show all other graph as well
x=x.add_legend()# To show legend
```

```python
x=sns.PairGrid(data=iris, vars=['sepal_length','petal_length']) # To show relation between specific coloumns
x=x.map(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data=iris, hue='species',vars=['sepal_length','petal_length']) # To show relation between specific coloumns #Adding Hue
x=x.map(plt.scatter)
x=x.add_legend()
```

```python
x=sns.PairGrid(data=iris, hue='species',vars=['sepal_length','petal_length']) # To show relation between specific coloumns #Adding Hue
x=x.map_diag(plt.hist)## making diaganle graph as historgraph
x=x.map_offdiag(plt.scatter)##showing other graph as scatter
x=x.add_legend()
```

```python
#To interchange x and y , specific which colums you need on x and y by following
x=sns.PairGrid(data=iris, hue='species',x_vars=['petal_length','petal_width'], y_vars=['sepal_length','sepal_width']) 
x=x.map(plt.scatter)##showing other graph as scatter
x=x.add_legend()
```

```python
# Can also show x and y axis as per your need
x=sns.PairGrid(data=iris, hue='species',x_vars=['petal_length'], y_vars=['sepal_length','sepal_width']) 
x=x.map(plt.scatter)##showing other graph as scatter
x=x.add_legend()
```

```python
# To show different types of plat accors diagnal use 
x=sns.PairGrid(data=iris, hue='species') 
x=x.map_diag(plt.hist)##specificy diagonal graph type
x=x.map_upper(sns.kdeplot)###specificy upper diagonal graph type
x=x.map_lower(plt.scatter)###specificy lower diagonal graph type
x=x.add_legend()
```

```python
# Here we will change marker style for hue .
x=sns.PairGrid(data=iris, hue='species', hue_kws={'marker':['D','s','+']}) #Since we have 3 hue variable. We will pass dic with list
x=x.map(plt.scatter, s=30)
x=x.add_legend()
```

```python
flights=sns.load_dataset("flights")
```

```python
flights
```

```python
flights=flights.pivot('month','year','passengers')
flights
```

```python
sns.heatmap(flights) # heat map showing frequency of passenger based on ordered months and year
```

```python
sns.clustermap(flights)# Cluster map showing similar/cluster of passenger based on both year and month
```

```python
sns.clustermap(flights, col_cluster=False)# Shoing only similar/cluster of passenger based on rows only. 
#rows are month based on pivot table
#there rows/month are unordered
```

```python
sns.clustermap(flights, row_cluster=False)# Shoing only similar/cluster of passenger based on col only. 
```

```python
sns.clustermap(flights, row_cluster=False, cmap="coolwarm", linewidth=1)
#chaining line width and color
```

```python
sns.clustermap(flights, row_cluster=False, cmap="coolwarm", linewidth=1, figsize=(8,8))
#chaingn size
```

```python
sns.clustermap(flights, cmap="coolwarm", linewidth=1, figsize=(8,8), standard_scale=0)
#standardzing data accross  row 0 row and 1 columns
```

```python
sns.clustermap(flights, cmap="coolwarm", linewidth=1, z_score=0)
#standardzing data accross  row #0 row and 1 colomuns
```
