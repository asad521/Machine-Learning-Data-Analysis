```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
##Box have satuation and dots have alpha
```

```python
tips=sns.load_dataset("tips")
```

```python
tips
```

```python
x=sns.FacetGrid(tips, row='smoker', col='time')# time and smoker is categorilca variable. This will make possible combination of
#these categorical variable
x=x.map(plt.hist, "total_bill") #Now based on above graph, this will show the histrogram of total bill based on above numerical values
#Basically Facet grid is used to make graph based on categorical variables and then plot numerical data based on thoses graph
```

```python
x=sns.FacetGrid(tips, row='smoker', col='time')# time and smoker is categorilca variable. This will make possible combination of
#these categorical variable
x=x.map(plt.scatter, "total_bill", "tip")
# Facet Grid first draw possible graph based on conditions of caterogical variable
#Then he simply draw scatter plot of total, bill and tip based on thoses conditions
```

```python
x=sns.FacetGrid(tips, row='smoker', col='time')# time and smoker is categorilca variable. This will make possible combination of
#these categorical variable
x=x.map(sns.regplot, "total_bill", "tip")
# Facet Grid first draw possible graph based on conditions of caterogical variable
#Then he simply draw regression plot based on above  based on thoses conditions
```

```python
x=sns.FacetGrid(tips, col='smoker', hue='sex')# using only one variable and hue# only one condition/categoircal varible
#these categorical variable
x=x.map(plt.scatter, "total_bill", "tip").add_legend() # legend is added 
```

```python
x=sns.FacetGrid(tips, col='day', size=4, aspect=.8)# using only one variable and hue# only one condition/categoircal varible
#these categorical variable
x=x.map(sns.boxplot, "total_bill", "sex").add_legend() # legend is added 
```
