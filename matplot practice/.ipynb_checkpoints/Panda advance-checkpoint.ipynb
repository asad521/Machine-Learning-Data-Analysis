```python
df=pd.read_csv("weatherpp.csv")
```

```python
df
```

```python
df1=df.pivot_table(index="city", columns="date") #it will shows average temperatre based on column and city
df1
```

```python
df1=df.pivot_table(index="city", columns="date", aggfunc='sum') #it will show sum  temperatre based on column and city
df1
#aggfunc= sum,diff, and many
```

```python
df1=df.pivot_table(index="city", columns="date", margins=True) #it will show colum and row wise mean
df1
```

```python
df=pd.read_csv("weatherppp.csv")
df['date']=pd.to_datetime(df['date'])
df
```

```python
df1=df.pivot_table(index=pd.Grouper(freq="m", key="date"), columns="city") #groupper will group based on date frequency and show average
df1
```
