```python
#OneHotEcoding By sklearn Library
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohc=OneHotEncoder(sparse=False)
```

```python
ohc_arr=ohc.fit_transform(tips_df[['sex','smoker','day','time']])
```

```python
pd.DataFrame(ohc_arr, columns=['sex_Male', 'sex_Female', 'smoker_Yes',
       'smoker_No', 'day_Thur', 'day_Fri', 'day_Sat', 'day_Sun', 'time_Lunch',
       'time_Dinner'])
```

```python
ohc_arr=ohc.fit_transform(tips_df[['sex','smoker','day','time']])
```

```python
ohc=OneHotEncoder(sparse=False,drop='first')
```

```python
ohc_arr=ohc.fit_transform(tips_df[['sex','smoker','day','time']])
ohc_arr
```

```python
a=pd.DataFrame(ohc_arr, columns=['male', 'smoker', 'Sun',
       'Thrd', 'Fri', 'time'])
a
```

```python
final_df=tips_df.drop(['sex','smoker','day','time'],axis=1)
final_df
```

```python
final_df=pd.concat([final_df,a], axis=1)
final_df
```
