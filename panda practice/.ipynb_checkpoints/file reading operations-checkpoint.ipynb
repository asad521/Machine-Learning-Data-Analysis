# excel file reading


```python
df=pd.read_excel("stock_data.xlsx", skiprows=1)  #It will skip first row, and make 2nd row as header/index.
df
#This is also same to following.#It will skip first row, and make 2nd row as header/index.
df=pd.read_excel("stock_data.xlsx", header=1)
df
# However if you do not have header then do following. It will generate custom columns with numerbs
df=pd.read_excel("stock_data.xlsx", header=None)
df
df=pd.read_excel("stock_data.xlsx", header=None, names=['tickers','eps','revenue','price','people'])  #If you want to generate customs columns
df
```

```python
df=pd.read_excel("stock_data.xlsx", header=[0,1], Sheet='Book1' ) #This file has two header/columns/index. Therefore, header list is passed
df
#Inner header is bigger , upper is lower

```

```python
# Better to use replace.
#Converter Specified values into na when opening file
df=pd.read_excel("stock_data.xlsx", na_values=["not available","n.a."])# This will give convert all [listed value] into na
df
```

```python
# Better to use replace.

#But what if you want convert some columns values and no others
#Then you need to pass dictionary
#Key are columns. They values are that element you want to changee
df=pd.read_excel("stock_data.xlsx", na_values={
                "eps": ['not available'],
                "revenue":[-1.0],
                "people":['not available',"n.a."],
                "price":['not available',"n.a."]
 }
 )
```

```python
# or you can pass a function 
#In above example you are converting unwanted data to Nan values,
#If you want to convert or assing unwanted data to something else
#Then you will useconverter function and define a function
def conversion(a):
    if a=="not available":
        return "kING"
    else:
        return a
        
df=pd.read_excel("stock_data.xlsx", converters={
                "eps": conversion,
})
df
```
