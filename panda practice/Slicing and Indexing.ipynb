```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
df.columns
```

### Raw Selection. (always provide range)


```python
df[1:9]
# df[[1,3,4]]# Not possible here
```

```python
# df[start point:end point] #it will show end-1 row.
df[0:10:2]  # 0 to 10 rows with step size 2
df[10:0:-2] #10 to zero rows with step size 2  
#One thing to note that it will always start from start and go to end point irrespective of direction.
```

### Columns Selection ( write column name)

```python
df['total_bill']
```

```python
df[['total_bill','sex']]
```

## row and column selection (raws are always in range)

```python
df[1:9]['day']
```

```python
#multiple raw and columns #Pass the list
df[1:8][['total_bill','day']]
```

# # row and column selection in  one line.(pd.loc[row range, column name])
## Only work for labels. 
## row range or row list is acceptable


```python
df.loc[1:5,['total_bill','sex','day']]
```

```python
df.loc[[1,4,5],['total_bill','sex','day']]
```

# # row and column selection in  one line.(pd.iloc[row range, column name])
## will only work when columns and rows are have integer index.
## row range or list is acceptable.

```python
df.iloc[34:88,4:9]
#Here you need to provide integers
```
