```python
import pandas as pd
import seaborn as sns
# pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df=sns.load_dataset("tips")
```

```python
df
```

```python
df1=pd.melt(df, id_vars=["sex"])
df1
```

```python
df1=pd.melt(df, id_vars=["day"], value_vars=["sex","time"],var_name="Comp Col", value_name="All val")
df1
```

```python
df1=pd.melt(df, id_vars=["day",'sex',], value_vars=["total_bill","tip"],var_name="Comp Col", value_name="All val")
df1
```
