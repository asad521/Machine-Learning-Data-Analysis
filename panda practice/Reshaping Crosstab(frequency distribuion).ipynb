```python
import pandas as pd
import numpy as np
#will show the frequency between 2 columns.
#Better both be categorical
```

```python
df=pd.read_excel("survey.xls", Sheet='Book1')
```

```python
pd.crosstab(df.Sex,df.Handedness)# showing the frequency of Sex columns categori(male,female) in terms of handiness
```

```python
 pd.crosstab(df.Sex,df.Handedness, margins=True)# adding summary columns
```

```python
 pd.crosstab(df.Sex,[df.Handedness,df.Nationality], margins=True)
```

```python
 pd.crosstab(df.Sex,[df.Handedness,df.Nationality], margins=True)#checking sex(male &female) based on left and rights and counties
```

```python
 pd.crosstab(df.Sex,[df.Nationality,df.Handedness], margins=True)#checking sex(male &female)
#     based on left and rights and counties.
```

```python
pd.crosstab([df.Sex,df.Nationality],[df.Handedness], margins=True)
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Handedness], margins=True)
```

```python
#noramlized
pd.crosstab(df.Sex,df.Handedness, margins=True)
```

```python
#noramlized
pd.crosstab(df.Sex,df.Handedness, normalize='index')# if you want to check the frequency (2/5 for fl and 5/7 for ml)rowwise
```

```python
pd.crosstab(df.Sex,df.Handedness, normalize='columns')# if you want to check the frequency (2/7 for fl and 5/7 for ml) colwise
```

```python
pd.crosstab(df.Sex,df.Handedness, normalize='all')# if you want to check the frequency (2/12 for fl and 5/12 for ml) marginwise
# total number of student
```

# Aggregate

```python
# Aggrete function will change to cross tab to behave to show as per given aggfunc
# Now below left and  right columsn are now showing frequncy of left or right
# it showing average age(aggfunc=mean) of left handed and right handed famles and males
```

```python
pd.crosstab(df.Sex,df.Handedness, values=df.Age, aggfunc=np.mean)
```
