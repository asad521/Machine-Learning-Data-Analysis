# DataFrame.pivot(self, index=None, columns=None, values=None)

```python
# Important point
# index=col string --> index columns must contain all unique values. categorical columns should be selected
# columns=col string-->categorical columns should be selected
#values=num col string or list of num col string-->colouns contain numerical values that you want to see 

```

### DataFrame.pivot_table(self, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False,dropna=True, margins_name='All', observed=False)

```python
#index=string col. Do not need to be unique.categorical columns should be selected
#columns=categorical columns should be selected
#values=num col string or list of num col string-->colouns contain numerical values whose statistic you want to see 
# aggfunc= String or list."mean", "sum","prod" All function supported by np or panda. 
#margin= it is summary column and row.
#dropna= Iff all values of col is na , then it will drop it
```

```python
import pandas as pd
import seaborn as sns
```

```python

df=sns.load_dataset("tips")
```

```python
df
```

```python
df.pivot_table(index="day", columns="sex", values=["tip","total_bill"])
```

```python
df.pivot_table(index="day", columns="sex", values=["tip","total_bill"], aggfunc=['sum','mean', 'prod'])
```

```python
df.pivot_table(index="day", columns="sex", values=["tip","total_bill"], aggfunc=['sum','mean', 'prod'], margins=True)
```

```python
df.pivot_table(index="tip", columns="sex")
```

```python
df.pivot_table(index="tip", columns="sex", fill_value=0)
```

```python
df.pivot_table(index="tip", columns="sex", dropna=True)
```
